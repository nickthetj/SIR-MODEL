```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# DEFINE CONSTANTS
a = 0.45 # infection rate
b = 0.1 # recovery rate
c = 0.2 # vax rate

# FUNCTION TO RETURN DERIVATIVES AT T
def f(y,t):
    
    S, I, R, V = y # get previous values of S, I, R and store them in array y

    d0 = -a*S*I - c*S # derivative of S(t)
    d1 = a*S*I - b*I # derivative of I(t)
    d2 = b*I # derivative of R(t)
    d3 = c*S # derivative of V(t)

    return [d0, d1, d2, d3]

# INITIAL VALUES OF EACH FUNCTION 
S_0 = 1
I_0 = 3.125/(10**6)
R_0 = 0
V_0 = 0

y_0 = [S_0,I_0,R_0,V_0]

t = np.linspace(start=1,stop=120,num=120)
y = odeint(f,y_0,t)

S = y[:,0]
I = y[:,1]
R = y[:,2]
V = y[:,3]

RV = y[:,2] + y[:,3]
plt.figure()
plt.plot(t,S,'r',label="% of Susceptibles")
plt.plot(t,I,'b',label="% of Infected")
plt.plot(t,R,'g',label="% of Recovered")
plt.plot(t,V,'purple',label="% of Vaccinated ")


plt.legend()
plt.show()
```
